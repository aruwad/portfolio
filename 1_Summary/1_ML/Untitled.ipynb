```python
import winsound
import time

# Frequencies (in Hz) for each note
notes = {
    "도": 261,  # C4
    "레": 293,  # D4
    "미": 329,  # E4
}

# Melody: "미 레 도 레 미 미 미"
melody = ["미", "레", "도", "레", "미", "미", "미"]

# Duration per note (milliseconds)
duration = 300

# Play melody
for note in melody:
    winsound.Beep(notes[note], duration)
    time.sleep(0.05)  # slight pause between notes
```
