# 1. Basics.

## Data.

```python
from datasets import load_dataset

ds = load_dataset('imdb')

train_ds = ds['train']
test_ds  = ds['test']

train_ds = train_ds.select(range(50))
```

```python
print("This is the original text.")
train_ds['text'][0]
```

## Tokenize.

```python
# Initialization.
from transformers import AutoTokenizer

checkpoint = 'distilbert-base-uncased'
tokenizer  = AutoTokenizer.from_pretrained(checkpoint)

# Tokenize method.
def tokenize(sample):
    tokenized = tokenizer(sample['text'])
    
    return tokenized

# Tokenize.
train_ds_tokenized = train_ds.map(tokenize)
```

```python
print("Tokenized: Default.")
print(f"Length = {len(train_ds_tokenized['input_ids'][0])}")
print(train_ds_tokenized['input_ids'][0])
```

```python
for idx, sample in enumerate(train_ds_tokenized):
    if len(sample['input_ids']) < 256:
        print(idx)
        break
```

```python
# truncation and padding.
def tokenize(sample):
    tokenized = tokenizer(sample['text'],
                          max_length=256,
                          truncation=True, 
                          padding='max_length')
    return tokenized
```

```python
train_ds_tokenized = train_ds.map(tokenize)
```

```python
# truncated.
print(len(train_ds_tokenized['input_ids'][0]))
print(train_ds_tokenized['input_ids'][0])
```

```python
# padded.
print(len(train_ds_tokenized['input_ids'][2]))
print(train_ds_tokenized['input_ids'][2])
```

```python
# tokenize multiple columns.
from copy import deepcopy

train_ds_mul = deepcopy(train_ds)

train_ds_mul = train_ds_mul.map(lambda x: {'text_cp': x['text']})
```

```python
def tokenize(sample):
    tokenized = tokenizer([sample['text'], sample['text_cp']],
                          max_length=256,
                          truncation=True,
                          padding="max_length")
    return tokenized
```

```python
train_ds_mul_tokenized = train_ds_mul.map(tokenize)
```

```python
train_ds_mul_tokenized['input_ids'][0][0]   # ['text']
train_ds_mul_tokenized['input_ids'][0][1]   # ['text_cp']
```

```python
# Stride.
def tokenize(sample):
    tokenized = tokenizer(sample['text'],
                          max_length=128,
                          truncation=True,
                          padding='max_length',
                          stride=8,
                          return_overflowing_tokens=True   # stride only works with this=True!
                         )
    return tokenized
```

```python
train_ds_tokenized_stride = train_ds.map(tokenize)
```

```python
# One sample contains 3 elements now!
print(len(train_ds_tokenized['input_ids'][0]))
print(len(train_ds_tokenized_stride['input_ids'][0]))  # 0~128, 125~253, 250~363.
```

```python
# Now shape should be changed for training. 'input_ids': (50,3,128) and 'label': (50,1).
print(len(train_ds_tokenized_stride['input_ids']))
print(len(train_ds_tokenized_stride['input_ids'][0]))
print(len(train_ds_tokenized_stride['input_ids'][0][0]))
print(len(train_ds_tokenized_stride['label']))
```

```python
def duplicate_labels(sample):
    return {'label': [sample['label']] * len(sample['input_ids'])}
```

```python
# Duplicate label so that it has (50,3,1).
# 1 => [1,1,1].
train_ds_tokenized_stride = train_ds.map(tokenize)

train_ds_tokenized_stride = train_ds_tokenized_stride.map(
    duplicate_labels
)
```

```python
train_ds_tokenized_stride['label'][0]    # (50,3,1)!
```
